```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import kagglehub

# Download latest version
path = kagglehub.dataset_download("arshid/iris-flower-dataset")

print("Path to dataset files:", path)
```

```python
iris = load_iris()
x=iris.data
y=iris.target
```

```python
scaler=StandardScaler()
x_scaler=scaler.fit_transform(x)
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_scaler)
```

```python
plt.figure(figsize=(18,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=y,cmap='viridis',edgecolors='k',s=100)
plt.title('PCA of Iris Dataset (2 Component)')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.colorbar(label='Target Class')
plt.show()
```

```python
print("Explained Variance Ratio of Each component:")
print(pca.explained_variance_ratio_)
```

```python
pca=PCA(n_components=3)
x_pca=pca.fit_transform(x_scaler)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(18,8))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x_pca[:,0],x_pca[:,1],x_pca[:,2],c=y,cmap='viridis',edgecolors='k',s=100)
ax.set_title('PCA of Iris Dataset (3 Component)')
ax.set_xlabel(' Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.show()
```
