```python
import pandas as pd
```

```python
url='/content/titanic_clean.csv'
data=pd.read_csv('/content/titanic_clean.csv')
```

```python
summary=data.describe()
```

```python
print(summary)
```

```python
print(data.describe(include='all'))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.hist(data['ticket'].dropna(), bins=100, edgecolor='black',color='skyblue')
plt.title('ticket Distribution')
plt.xlabel('ticket')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
data['pclass'].value_counts().plot(kind='bar',color='lightcoral')
plt.title('passenger class Distribution')
plt.xlabel('class')
plt.ylabel('count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(data['age'].dropna(),patch_artist=True,
boxprops=dict(facecolor='lightblue'))
plt.title('box plot of age')
plt.xlabel('age')
plt.xlabel('age')
plt.show()
```

```python
sns.pairplot(data[['age','fare','sibsp','parch']].dropna())
plt.show()
```

```python
correlation_matrix=data[['age','fare','sibsp','parch']].corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',cbar=True)
plt.title('Survival count')
plt.xlabel('survived')
plt.ylabel('count')
plt.show()
```

```python
sns.countplot(x='survived', data=data, palette='Set2') # Capitalize 'Set2'
plt.title('Survival count')
plt.xlabel('survived')
plt.ylabel('count')
plt.show()
```

```python
sns.violinplot(x='pclass', y='age', data=data, palette='muted')
plt.title('age Distribution by passenger class')
plt.show()
```
